```python
def date_converter():
    m = ['January','February','March','April','May','June','July','August','September','October','November','December']
    n = input('Enter a date in the format \'mm/dd/yy\': ')
    flag = 0
    if n.count('/') == 2:
        n = n.split('/')
        if (1 <= int(n[0]) <= 12) and (1 <= int(n[1]) <= 31):
            mm, dd, flag = int(n[0]) - 1, int(n[1]), 1
        else:
            print('Invalid value')
        if 99 >= int(n[2]) >= 23 and flag == 1:
            yy = int('19' + n[2])
            print('Date: {} {}, {}'.format(m[mm], dd, yy))
        elif 0 <= int(n[2]) <= 9 and flag == 1:
            yy = int('200' + n[2])
            print('Date: {} {}, {}'.format(m[mm], dd, yy))
        elif 10 <= int(n[2]) <= 22 and flag == 1:
            yy = int('20' + n[2])
            print('Date: {} {}, {}'.format(m[mm], dd, yy))
        elif 1900 <= int(n[2]) <= 2099 and flag == 1:
            yy = int(n[2])
            print('Date: {} {}, {}'.format(m[mm], dd, yy))
        elif (int(n[2]) >= 100 or int(n[2]) < 0) and flag == 1:
            print('Invalid value')
    else:
        print('Invalid value')
date_converter()
```

```python
def bmm(m, n):
    bmm, large = 1, m if m > n else n
    for i in range(2,large + 1):
        if n%i == 0 and m%i == 0:
            bmm = i
    return bmm
    
def fraction_simplifier():
    n = input('Enter a fraction in format \'n/m\': ')
    if n.count('/') == 1:
        m = n.index('/')
        x, y = int(n[ :m]), int(n[m+1: ])
        z = bmm(x,y)
        x, y = x/z, y/z
        print('{}/{}'.format(int(x), int(y)))
    else:
        print('Invalid Input')
fraction_simplifier()
```
